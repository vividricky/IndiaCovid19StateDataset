```python
Dose1=pd.DataFrame()
Dose2=pd.DataFrame()
DoseTotal=pd.DataFrame()
for i in range(len(names)):
    data=pd.read_csv('csv/%s' %names[i],names = newcolmuns)
    data=data[1:]
    data=data.drop('Unnamed', 1)
    data=data.T
    data.columns=state
    Dose1temp=data[1:2]
    Dose1temp.index=pd.Series(["%s" %(names[i][0:8])])

    Dose1= Dose1.append(Dose1temp)
    
    Dose2temp=data[2:3]
    Dose2temp.index=pd.Series(["%s" %(names[i][0:8])])
    Dose2= Dose2.append(Dose2temp)
    
    DoseTotaltemp=data[3:4]
    DoseTotaltemp.index=pd.Series(["%s" %(names[i][0:8])])
    DoseTotal= DoseTotal.append(DoseTotaltemp)
```
